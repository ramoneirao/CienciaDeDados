### Importando as bibliotecas necessárias

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Carregar os dados
df = pd.read_csv('ecommerce_preparados.csv')
print(df.columns)
```

```python
# Histograma
plt.figure(figsize=(8,6))
sns.histplot(df['Nota'], bins=20, kde=True)
plt.title("Distribuição das Notas")
plt.xlabel("Nota")
plt.ylabel("Frequência")
plt.show()
```

```python
# Gráfico de dispersão
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['N_Avaliações'], y=df['Nota'])
plt.title("Correlação entre Número de Avaliações e Nota")
plt.xlabel("Número de Avaliações")
plt.ylabel("Nota")
plt.show()
```

```python
# Mapa de calor
plt.figure(figsize=(8,6))
correlation_matrix = df[['Nota', 'N_Avaliações', 'Desconto', 'Qtd_Vendidos_Cod']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Mapa de Calor das Correlações")
plt.show()
```

```python
# Diminuindo a quantidade de gêneros para uma melhor vizualização
df['Gênero'] = df['Gênero'].apply(lambda x: x if x in ['Masculino', 'Feminino', 'Sem gênero'] else 'Sem gênero')

# Gráfico de barra
plt.figure(figsize=(8,6))
sns.barplot(x=df['Gênero'].value_counts().index, y=df['Gênero'].value_counts().values)
plt.title("Distribuição de Produtos por Gênero")
plt.xlabel("Gênero")
plt.ylabel("Contagem")
plt.show()
```

```python
# Gráfico de pizza
plt.figure(figsize=(8,6))
df['Gênero'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Distribuição de Produtos por Gênero")
plt.show()
```

```python
# Gráfico de densidade
plt.figure(figsize=(8,6))
sns.kdeplot(df['N_Avaliações'], fill=True)
plt.title("Densidade do Número de Avaliações")
plt.xlabel("Número de Avaliações")
plt.ylabel("Densidade")
plt.show()
```

```python
# Gráfico de regressão
plt.figure(figsize=(8,6))
sns.regplot(x=df['N_Avaliações'], y=df['Nota'])
plt.title("Relação entre Número de Avaliações e Nota")
plt.xlabel("Número de Avaliações")
plt.ylabel("Nota")
plt.show()
```
